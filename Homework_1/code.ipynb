```python
import pandas as pd
import altair as alt
```

#### Read data

```python
df = pd.read_csv('GLB.Ts_dSST.csv', skiprows=[0])
df.head()
```

#### Data Processing

```python
temp_df = df.copy()
temp_df.drop(['Year', 'J-D','D-N', 'DJF', 'MAM', 'JJA','SON'], axis = 1, inplace = True)
temp_df.head()
```

```python
df['average_temp'] = temp_df.mean(axis = 1)
# df.head()
```

```python
df1 = df.copy()
df1.drop(['J-D','D-N', 'DJF', 'MAM', 'JJA','SON', 
          'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
         , axis = 1, inplace = True)
df1.head()
```

## Chart

```python
main_chart = alt.Chart(df1).mark_point(
    
            color = '#404040', fill = '#f06400',
            size = 90, opacity = 1, strokeWidth = 1
            ).encode(
    
    x = alt.X(field='Year', type = 'quantitative', title='',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(format = 'k', tickCount = 13, labelFontSize=13, labelColor='#b3b3b3')),
    
    y = alt.Y(field = 'average_temp', type = 'quantitative', title='',
              scale = alt.Scale(zero = False),
              axis = alt.Axis(tickCount = 10, labelFontSize=13, labelColor = '#b3b3b3'))
)


zero_line = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='#A9A9A9').encode(y='y')


title_upper_text = alt.Chart({'values':[{'x': 1883, 'y': 0.93}]}
                            ).mark_text(
                                text='Annual Global Surface Temperature,', 
                                align='left', fontSize=15, fontWeight=800
                            ).encode(x='x:quantitative', y='y:quantitative')

title_bottom_text = alt.Chart({'values':[{'x': 1883, 'y': 0.87}]}
                             ).mark_text(
                                text='Relative to Late 19th Century Average',
                                align='left', fontSize=15, fontWeight=800
                             ).encode( x='x:quantitative', y='y:quantitative')


right_upper_text = alt.Chart({'values':[{'x': 2017, 'y': 0.13}]}
                            ).mark_text(
                                text='HOTTER THAN THE', align='right', 
                                fontWeight=500
                            ).encode( x='x:quantitative', y='y:quantitative')

right_middle_text = alt.Chart({'values':[{'x': 2017, 'y': 0.07}]}
                            ).mark_text(
                                text='1880-1899 AVERAGE', align='right',
                                fontWeight=500
                            ).encode( x='x:quantitative', y='y:quantitative')

right_bottom_text = alt.Chart({'values':[{'x': 2017, 'y': -0.07}]}
                            ).mark_text(
                                align='right',
                                text='COLDER',
                                fontWeight=500
                            ).encode( x='x:quantitative', y='y:quantitative')


years = [1904, 1944, 1998, 2014, 2015, 2016, 2017]
years_var = [0, 0, 0, 0, 0, 0, 0]
    
for i in range(len(years)):
    years_var[i] = alt.Chart({'values':[{'x': years[i]-1, 
                                         'y': df1._get_value(df1.index[df1['Year']==years[i]].tolist()[0], 'average_temp')}]}
                        ).mark_text(
                            align='right', text=str(years[i]),
                            fontWeight=500, fontSize=15,
                            ).encode( x='x:quantitative', y='y:quantitative')


alt.layer(main_chart, zero_line, 
              title_upper_text, title_bottom_text, 
              right_upper_text, right_middle_text, right_bottom_text,
              years_var[0], years_var[1], years_var[2], years_var[3], 
              years_var[4], years_var[5], years_var[6]
         ).properties(width = 900, height = 500, background = '#F9F9F9', padding = 25
         ).configure_axis(ticks = False, domain = False)
```
